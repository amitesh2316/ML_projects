```python
import pandas as pd
import numpy as np
from pandas import read_csv
from pandas.plotting import scatter_matrix
from matplotlib import pyplot
import seaborn as sns
df.hist()
pyplot.show()
```

```python
from scipy.stats import zscore
z = np.abs(zscore(df))
z
```

```python
threshold = 3
print(np.where(z>3))
```

```python
df_new = df[(z < 3).all(axis=1)]
```

```python
print(df.shape)

print(df_new.shape)
```

```python
x = df_new.iloc[:,0:8]
```

```python
x
```

```python
x.shape
```

```python
x = x.reshape(-1,1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x = pd.DataFrame(x)
```

```python
y = df_new.iloc[:,-1]
```

```python
y
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=33,random_state=47) 
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
lm.intercept_
```

```python
lm.score(x_train,y_train)
```

```python
pred = lm.predict(x_test)
```

```python
print("predicted CMS",pred)
print("actual CMS",y_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,pred))
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,pred)
```

```python
MSE = np.square(np.subtract(y_test,pred)).mean()
MSE
```

```python
max_r_score = 0
for r_state in range(42,100):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=33,random_state=r_state)
    lm = LinearRegression()
    lm.fit(x_train,y_train)
    pred = lm.predict(x_test)
    r2_score = r2_score(y_test,pred)
    if r2_score>max_r_score:
        max_r_score=r2_score
        final_r_state=r_state

print("max r2 score corresponding",final_r_state,"is",max_r_score)
    
    
```
